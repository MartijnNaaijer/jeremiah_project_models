```python
import os
```

```python
model_keys = ['BEREL MEAN', 'BEREL CLS', 'OUR MEAN', 'OUR CLS']
```

```python
file_names = [file for file in os.listdir() if file.endswith('txt')]
```

```python
def get_hebrew_text(line):
    return line.split(') ')[1]

def get_section(line):
    section_joined = line.split(') ')[0]
    #print(section_joined.split(', '))
    bo_ch, clauses_str = section_joined.split(', (')
    clauses_str = clauses_str.strip(')')
    bo_ch = bo_ch.strip('(')
    bo, ch = bo_ch.split(', ')
    ch = int(ch)
    #print(clauses_str)
    clauses = tuple(int(cl) for cl in clauses_str.split(', '))
    #print(clauses)
    return bo, ch, clauses

def check_model_line(model_keys, line):
    for model_key in model_keys:
        if line.startswith(model_key):
            return model_key

    return None
        
```
